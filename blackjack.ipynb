```python
import random
import time
from IPython.display import clear_output

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return self.rank +' of '+ self.suit
    
class Deck:
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp = ''
        print('The deck has:')
        for x in self.deck:
            deck_comp = deck_comp + '\n' + str(x)
        return deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card_dealt = self.deck.pop()
        return single_card_dealt

class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
        
    def adjust_for_ace(self):
        if self.value > 21 and self.aces > 0:
            self.value -= 10
            self.aces -= 1
            
    def handcomp(self):
        hand_comp = self.cards[1] #returning the second card in the list, this second card is also the one which will appear facing upwards during player's turn
        return (str(hand_comp))[0]
        
class Chips:
    def __init__(self,bankroll):
        self.bankroll = bankroll  # This can be set to a default value or supplied by a user input
        self.bet = 0
        self.insurance = 0
        
    def win_bet(self):
        self.bankroll += self.bet
    
    def lose_bet(self):
        self.bankroll -= self.bet
        
    def insurancetriggered(self):
        self.bankroll = self.bankroll + (self.insurance * 4)
        
def take_bet():
    while True:
        try:
            chips.bet = int(input('How much would you like to bet? '))
        except:
            print('Please input an integer.')
        else:
            if chips.bet > chips.bankroll:
                print(f'You do not have sufficient chips to bet {chips.bet}')
            else:
                break
        
                
def hit(deck,hand):
    
    hand.add_card(test_deck.deal())
    hand.adjust_for_ace()

def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    while True:
        choice = input('Do you want to Hit or Stand? H/S: ')
        
        if choice[0].upper() == 'H':
            hit(deck,hand)
        
        elif choice[0].upper() == 'S':
            playing = False
        
        else:
            print('Pls input H/S only.')
            continue
        
        break

def show_some(player,dealer):
    
    print('Dealer has:')
    print('Hidden card')
    print(dealer.cards[1])
    
    print('\nPlayer has:')
    print(*player.cards, sep = '\n') #the asterisk * symbol is used to print every item in a collection (in this case its a list), sep = '\n' arguement print each item on a seperate line.
    print('\nTotal value is:',player.value, '\n')
    
def show_all(player,dealer):
    
    clear_output()
    print('Dealer has:')
    print(*dealer.cards, sep = '\n') 
    print('\nTotal value is:',dealer.value)
    
    print('\nPlayer has:')
    print(*player.cards, sep = '\n') 
    print('\nTotal value is:',player.value, '\n')
    pass

def player_busts(chips):
    chips.lose_bet()
    print('Player lost!')

def player_wins(chips):
    chips.win_bet()
    print('Player won!')
    
def dealer_busts(chips):
    chips.win_bet()
    print('Player won!')
    
def dealer_wins(chips):
    chips.lose_bet()
    print('Player lost!')

def insurance_triggered(chips):
    chips.insurancetriggered()
    print('Insurance triggered')
    
def push():
    print('It''s a tie!')
    

playing = True 
print("Welcome to Daniel's BlackJack")
test_deck = Deck()
test_deck.shuffle()

while True:
        try:
            amount = int(input('How many chips do you want to start with? '))
        except:
            print('Pls input an integer')
        else:
            break
        
chips = Chips(amount)
    
while True:
    
    if test_deck.deck == []:
        test_deck = Deck()
        test_deck.shuffle()
        
    player = Hand()
    player.add_card(test_deck.deal())
    player.add_card(test_deck.deal())

    dealer = Hand()

    dealer.add_card(test_deck.deal())
    dealer.add_card(test_deck.deal())
  
    
    # Prompt the Player for their bet
    print(f'You currently have {chips.bankroll} in chips')
    take_bet()
    
    # Show cards (but keep one dealer card hidden)
    clear_output()
    show_some(player,dealer)
    
    if dealer.handcomp() == 'A':
        chips.insurance = float(chips.bet/2)
        insurance_in_place = None
        if (chips.bankroll-chips.bet) > chips.insurance:
            while True:
                insurance = input(f'Do you want to purchase insurance for {chips.insurance}? Y/N: ')
                if insurance[0].upper() == 'Y':
                    insurance_in_place = True
                elif insurance[0].upper() == 'N':
                    insurance_in_place = False
                else:
                    print('Pls enter Y/N: ')
                    continue 
                break
        else:
            print('You do not have enough chips to buy insurance')
    
    while playing:  # recall this variable from our hit_or_stand function # while True since playing = True at the top
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(test_deck,player)
        clear_output()
        
        # Show cards (but keep one dealer card hidden)
        show_some(player,dealer)
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player.value > 21:
            player_busts(chips)
            break
        
    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    
    if player.value <= 21:
        while dealer.value < 17:
            print('Dealing card...')
            time.sleep(3)
            hit(test_deck,dealer)
            show_all(player,dealer)

        # Run different winning scenarios
        if dealer.value > 21:
            show_all(player,dealer)
            dealer_busts(chips)
            
        elif dealer.value == 21 and dealer.handcomp() =='A' and insurance_in_place == True and dealer.value > player.value:
            show_all(player,dealer)
            dealer_wins(chips)
            insurance_triggered(chips)
            
        elif dealer.value == 21 and dealer.handcomp() =='A' and insurance_in_place == True and dealer.value < player.value:
            show_all(player,dealer)
            player_wins(chips)
            insurance_triggered(chips)

        elif dealer.value == 21 and dealer.handcomp() =='A' and insurance_in_place == True and dealer.value == player.value:
            show_all(player,dealer)
            push()
            insurance_triggered(chips)
            
        elif player.value > dealer.value:
            show_all(player,dealer)
            player_wins(chips)
        

        elif dealer.value > player.value:
            show_all(player,dealer)
            dealer_wins(chips)

        else:
            show_all(player,dealer)
            push()
        
    # Inform Player of their chips total 
    print(f'You currently have {chips.bankroll} in chips')
    
    # Ask to play again
    if chips.bankroll == 0:
        print('You have ran out of chips to play.')
        
    while True:
        play_again = input('Do you want to play again? Y/N: ')

        if play_again[0].upper() == 'Y':
            if chips.bankroll <= 0:       
                while True:
                    try:
                        amount = int(input('How many chips do you want to start with? '))
                    except:
                        print('Pls input an integer')
                    else:
                        break  
                chips = Chips(amount)
            
        elif play_again[0].upper() == 'N':   
            print('Thanks for playing')
            break
    
        else:
            print('Pls enter Y/N.')
            continue
        break
            
    if play_again[0].upper() == 'Y':
        playing = True
        clear_output()
        
    elif play_again[0].upper() == 'N':
        break
```
